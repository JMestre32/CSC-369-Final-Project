```python
import json
```

```python
def extract_first_company_name(json_str):
    try:
        companies = json.loads(json_str)
        if companies:
            return companies[0]['name']
        else:
            return None
    except json.JSONDecodeError:
        return None

# Apply the function to the 'production_companies' column
df['first_company_name'] = df['production_companies'].apply(extract_first_company_name)

# Display the DataFrame with the new 'first_company_name' column
print(df[['production_companies', 'first_company_name']])
```

```python
def genre(json_str):
    try:
        companies = json.loads(json_str)
        if companies:
            return companies[0]['name']
        else:
            return None
    except json.JSONDecodeError:
        return None

# Apply the function to the 'production_companies' column
df['genre'] = df['genres'].apply(genre)

# Display the DataFrame with the new 'first_company_name' column
print(df[['genre', 'genres']])
```

```python
data = pd.read_csv('output.txt', sep=":", header=None)
data.columns = ["category", "value"]
data
```

```python
data['isCompany'] = data['category'].str.startswith('companyVec').astype(int)
data['isGenre'] = data['category'].str.startswith('genreVec').astype(int)
```

```python
data.head()
```

```python
genre_df = data[data["isGenre"] == 1]
company_df = data[data["isCompany"] == 1]
```

```python
sorted = genre_df.sort_values(by="value", ascending = False)
sorted_comp = company_df.sort_values(by = "value", ascending = False)
```

```python
df_compPop = pd.read_csv('output3.txt', sep=":", header=None)
df_compPop.columns = ["negligible", "compName", "popularity"]
df_compPop
```

```python
def getCompanyName(line):
  a = line.split(",")
  return a[0]
df_compPop['companyName'] = df_compPop['compName'].apply(getCompanyName)
```
